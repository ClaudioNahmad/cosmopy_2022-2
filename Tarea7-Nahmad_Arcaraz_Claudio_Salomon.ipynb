```python
k, t, r, theta, phi = symbols('k t r \\theta \\phi') #define simbolos, las dobles barras son para que se impriman en latex

a=Function('a')

chi = Coordinates('chi', [t, r, theta,phi]) #chi depende de t r theta y phi

metric = diag(-1,a(t)**2/(1-k*r**2),a(t)**2*r**2,a(t)**2*r**2*sin(theta)**2)

g = MetricTensor('g', chi, metric) #nombre de la metrica, coordenadas y la metrica
```

## 1. La métrica  de FLRW
### $ds^2 = g_{\mu \nu}dx^\mu dx^\nu = -dt^2 + a^2(t) \left(\dfrac{d \chi^2}{1- k \chi^2} + \chi^2 d \theta^2 + \chi^2 sin^2 \theta  d \phi^2 \right)$ 

```python
g(All,All)
```

```python
Gamma = Christoffel('Gamma', g) # Símbolos de Christofell
Gamma(-All, All, All)    # \Gamma^\mu_(\nu \sigma)
#Ga.components
```

```python
Rm = Riemann('Rm', g)
Rm(-All,All,All,All) 
```

```python
Ri = Ricci('Ri', g)   #Tensor de Ricci
Ri(All,All)
```

```python
Ri.scalar()
```

```python
G = Einstein('G', Ri)   #Tensor de Einstein
G(-All,All)
```
